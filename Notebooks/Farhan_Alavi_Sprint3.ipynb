### Making a linear model with Year, Long and Lat against Cases

```python
#Making a lin reg model from a few variables- Test 1
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

X1 = p[["Latitude", "Longitude", 'Year']]  # Independent variables
y1 = p['Total Cases']  # Dependent variable

# Splitting the data into training and testing sets
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2, random_state=42)

# Scaling
my_scaler=StandardScaler()
X1_train=my_scaler.fit_transform(X1_train)
X1_test=my_scaler.transform(X1_test)

# Initialize and fit the linear regression model
model = LinearRegression()
model.fit(X1_train, y1_train)

# Make predictions
y1_pred_train = model.predict(X1_train)
y1_pred_test = model.predict(X1_test)
```

```python
X1.corr()
```

```python
#Building an OLS model with Test 1
X1_withconstant = sm.add_constant(X1)
```

```python
# 1. Instantiate Model
myregression1 = sm.OLS(y1, X1_withconstant)

# 2. Fit Model
myregression1_results = myregression1.fit()

# Looking at the summary
myregression1_results.summary()
```

```python
# Assumptions - Calculate correlations between our independent variables.
corr1_df = X1.corr()

# Visualizing with seaborn
plt.figure(figsize=(10,7))
sns.heatmap(corr1_df.round(2), annot=True, vmax=1, vmin=-1, cmap='coolwarm')
plt.show()
```

```python
#Evaluating the scores from Test 1
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

# Evaluating the mean_absolute_error, mean_squared_error, and r2 numbers
train1_mae = mean_absolute_error(y1_train, y1_pred_train)
train1_mse = mean_squared_error(y1_train, y1_pred_train)
train1_r2 = r2_score(y1_train, y1_pred_train)

# Evaluating against the test data
test1_mae = mean_absolute_error(y1_test, y1_pred_test)
test1_mse = mean_squared_error(y1_test, y1_pred_test)
test1_r2 = r2_score(y1_test, y1_pred_test)

#printing the scores/results

print("Training MAE:", train1_mae)
print("Training MSE:", train1_mse)
print("Training R-squared:", train1_r2)
print("\nTest MAE:", test1_mae)
print("Test MSE:", test1_mse)
print("Test R-squared:", test1_r2)
```

```python
#plotting residuals
train1_residuals = y1_train - y1_pred_train
test1_residuals = y1_test - y1_pred_test

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(y1_pred_train, train1_residuals)
plt.title('Training Set Residuals')
plt.xlabel('Lat/Lon density by Year')
plt.ylabel('Total Cases')

plt.subplot(1, 2, 2)
plt.scatter(y1_pred_test, test1_residuals)
plt.title('Test Set Residuals')
plt.xlabel('Lat/Lon density by Year')
plt.ylabel('Total Cases')

plt.tight_layout()
plt.show()
```

```python
#Looking at the coeffs
coefficients1 = pd.DataFrame({'Variable': X1.columns, 'Coefficient': model.coef_})
print(coefficients1)
```

```python
coefficients1.set_index("Variable").sort_values("Coefficient", ascending=False).plot(kind="barh",)
```

```python
from scipy.stats import pearsonr

# Testing the correlations of each variable against 'Total Cases'
columns_to_test = p.columns[p.columns != 'Total Cases']

# Perform hypothesis test for each column
for col in columns_to_test:
    correlation, p_value = pearsonr(p[col], p['Total Cases'])
    print(f"Correlation between {col} and Total Cases: {correlation:.4f}, p-value: {p_value:.4f}")
```

A few of the P-values are below our 0.05 threshold so we can reject the null hypothesis for the following variables.
- Orange County
- San Diego
- Hawaiian Male

For the rest of them, we fail to reject.

Correlation with Total Cases:

Female Cases and Male Cases have very high positive correlations with Total Cases (0.9869 and 0.9915 respectively), suggesting a strong positive linear relationship.
Population, Males, Females, and other demographic groups also show moderate positive correlations with Total Cases.

However, many diseases and counties have negligible correlations with Total Cases, as indicated by correlation values close to zero.

P-values:

All the p-values are extremely small (0.000), indicating that the correlations are statistically significant.
Overall, the results suggest that there is a strong positive linear relationship between Female and Male Cases and Total Cases.
Additionally, moderate positive relationships exist between Total Cases and various demographic indicators like Population, Males, Females, etc. However, the correlations between diseases and Total Cases are generally low or negligible.
